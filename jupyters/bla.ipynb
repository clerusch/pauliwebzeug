```python
import pyzx as zx
from pyzx.pauliweb import compute_pauli_webs

c = zx.qasm("""
qreg q[2];
creg c[1];
cx q[0],q[1];
measure q[1] -> c[0];
""")
g = c.to_graph()
order, zwebs, xwebs = compute_pauli_webs(g)
zx.draw(g, labels=True)
print(zwebs)
```

```python
from qiskit import QuantumCircuit
from qiskit.qasm2 import dumps

qc = QuantumCircuit(2,1)
qc.cx(0,1)
qc.measure(1,0)
print(dumps(qc))
```
